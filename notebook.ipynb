# Predicting House Prices With A Regression Model

## Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading The CSV File to DataFrame

```python
train_file_path = 'train.csv'
train_data = pd.read_csv(train_file_path)
train_data.head()
```

## Data Cleaning

```python
train_data.info()
```

```python
train_data.drop( columns = [ 'Alley', 'PoolQC', 'Fence', 'MiscFeature', 'FireplaceQu', 'MasVnrType' ], 
                 axis = 1, 
                 inplace=True
                )

train_data = train_data.dropna()
train_data.info()
```

## Feature Engineering

```python
non_numeric_columns = train_data.select_dtypes(exclude=[np.number]).columns
print(non_numeric_columns)
```

```python
train_data['MSZoning'].value_counts()
```

## Model Build

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X = train_data.drop( columns = [ 'Id', 'SalePrice' ] )
y = train_data['SalePrice']
```

```python
model = RandomForestRegressor( random_state = 42 )
```

```python
model.fit( X, y )
```
